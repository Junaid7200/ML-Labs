```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2
y = 5
weights = np.linspace(-100, 100, 200)
loss = [(1 / (2 * 100)) * (y - w * x) ** 2 for w in weights]
plt.plot(weights, loss)
plt.show()
```

```python
x = 2
y = 1
b = 1
weights = np.linspace(-5, 5, 200)
loss = [1 / 2 * 100 * (y - (1 / (1 + np.exp(-w * x + b)))) ** 2 for w in weights]
plt.plot(weights, loss)
plt.show()
```
